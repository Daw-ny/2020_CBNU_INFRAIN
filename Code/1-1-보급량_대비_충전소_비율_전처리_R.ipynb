## 전기차 보급량 대비 충전소 설치비율 

```r
library(dplyr)
library(reshape2)
library(rgdal)
```

```r
jjirit = read.csv("01.광양시_충전기설치현황.csv",header=T)
```

```r
jjirit
```

```r
effort = read.csv("06.광양시_전기차보급현황(연도별,읍면동별).csv",header=T)
```

```r
effort <- effort %>%
            group_by(행정구역) %>%
            summarise(보급합=sum(보급현황))
## 행정구역 별 보급 현황의 합을 구함.
```

```r
eub <- c("광양읍","광양읍","중마동","광영동","금호동","광양읍","진월면","태인동",
        "광영동","광양읍","광영동","옥곡면","광양읍","중마동","중마동","광영동",
        "광영동","광양읍","중마동","중마동","광영동","진월면","진월면","진월면",
        "진월면","태인동","광양읍","광양읍","중마동","중마동","중마동","중마동",
        "중마동","중마동","중마동","중마동","중마동","중마동","중마동","광양읍",
        "광양읍","광양읍","광양읍","광양읍")
```

```r
data <- cbind(jjirit,eub)
```

```r
data[,c(4,6,11)]
```

```r
data_use <- data[,c(4,6,11)]
```

```r
good <- data_use %>%
        group_by(eub,급속.완속) %>%
        summarise(이용대수 = sum(이용대수)) %>% 
        dcast(eub~급속.완속)
## 급속, 완속 충전소를 나눔
```

```r
good
```

```r
good$급속[is.na(good$급속)] <- 0
```

```r
good
```

```r
effort_use <- effort %>% left_join(good,by=c("행정구역"="eub"))
```

```r
effort_use$급속[is.na(effort_use$급속)] <- 0
effort_use$완속[is.na(effort_use$완속)] <- 0
## 결측치 NA에 0을 대입
```

```r
effort_use <- effort_use %>% mutate(충전소합=급속+완속)
```

```r
effort_use$보급합[5]=0.1
effort_use
## 비율을 계산하기 위한 0을 0.1로 치환
```

```r
effort_use$급속[effort_use$급속==0] <- 0.1
effort_use$완속[effort_use$완속==0] <- 0.1
effort_use$충전소합[effort_use$충전소합==0] <- 0.1
## 비율을 계산하기 위한 0을 0.1로 치환
```

```r
effort_adm <- effort_use %>% mutate(급속보급필요율=보급합/급속,완속보급필요율=보급합/완속,
                                   총보급필요율=보급합/충전소합)
## 비율을 계산한 열을 추가
```

```r
need <- effort_adm[,c(1,6,7,8)]
## 필요한 열 선정
```

```r
need_use <- need %>% melt(id="행정구역")
```

```r
colnames(need_use) <- c("행정구역","구분","값")
```

```r
write.csv(need_use,"행정구역별_보급현황.csv",row.names=F)
## 행정구역별_보급현황.csv로 저장
```
