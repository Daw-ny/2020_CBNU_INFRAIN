```python
import folium
import json
import geopandas as gpd
import folium.plugins
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
```

```python
last = pd.read_csv("최종선정.csv")
last = last.iloc[:,1:4]
last_완속 = last[last['설치여부']=="완속"]
last_급속 = last[last['설치여부']=="급속"]
```

```python
m = folium.Map(
    location=[34.9406968,127.6958882],
    zoom_start=12
)
```

```python
for i in last_완속.index:
    # 행 우선 접근 방식으로 값 추출하기
    lat = last_완속.iloc[i, 2]
    lng = last_완속.iloc[i, 1]
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng],icon=folium.Icon(color='green',icon='info-sign'))
    marker.add_to(m)
```

```python
for i in last_완속.index:
    # 행 우선 접근 방식으로 값 추출하기
    lat = last_급속.iloc[i, 2]
    lng = last_급속.iloc[i, 1]
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng],icon=folium.Icon(color='blue',icon='info-sign'))
    marker.add_to(m)
```

```python
jjirit = pd.read_csv("01.광양시_충전기설치현황.csv")
place = jjirit[["충전소명","lon","lat"]]
```

```python
for i in place.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = place.iloc[i, 0]
    lat = place.iloc[i, 2]
    lng = place.iloc[i, 1]
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng], popup=name,icon=folium.Icon(color='red',
                                                                 icon='flag'))
    marker.add_to(m)
```
