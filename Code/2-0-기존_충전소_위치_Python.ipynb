```python
import folium
import json
import geopandas as gpd
import folium.plugins
import numpy as np
import pandas as pd
```

```python
jjirit = pd.read_csv("01.광양시_충전기설치현황.csv")
place = jjirit[["충전소명","lon","lat"]]
```

```python
GwangYang = folium.Map(location=[34.9406968,127.6958882], zoom_start=11)
```

```python
for i in place.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = place.iloc[i, 0]
    lat = place.iloc[i, 2]
    lng = place.iloc[i, 1]
    
    # 추출한 정보를 지도에 표시
    marker = folium.Marker([lat,lng], popup=name,icon=folium.Icon(color='red',
                                                                 icon='flag'))
    marker.add_to(GwangYang)
```
