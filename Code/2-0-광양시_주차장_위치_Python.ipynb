```python
import folium
import json
import geopandas as gpd
import numpy as np
import pandas as pd
```

```python
park = pd.read_csv("02.광양시_주차장_공간정보.csv")
park_1 = park[park["면적"]>0]
park_2 = park_1[park_1["구분"] !="임시"]
park_p = park_2[["구분", "면적", "주차면수", "주소", "lon", "lat"]]
park_pro = park_p[['lon', 'lat']]
```

```python
trans = gpd.read_file("26.광양시_도시계획(교통시설).geojson")
```

```python
map_all_park = folium.Map([34.963,127.5683], zoom_start=11)
for i in park.index:
    lat = park.loc[i, 'lat']
    lon = park.loc[i, 'lon']
    title = park.loc[i, '주차장명칭']
    Circle = folium.Circle([lat, lon], tooltip = title, radius=15, color='brown')
    Circle.add_to(map_all_park)
```

```python
trans_admin = trans[(trans['DGM_NM']!="일반철도")&(trans['DGM_NM']!="여객자동차터미널")&(trans['DGM_NM']!="자동차운전학원")&(trans['DGM_NM']!="일반철도(진상역)")&(trans['DGM_NM']!="궤도")&(trans['DGM_NM']!="항만시설")]
```

```python
folium.GeoJson(trans_admin,
               name='Parking_Station',
               style_function=lambda feature:{
                   'color': 'brown',     #border color for the color fills
                   'weight': 1
               }
              ).add_to(map_all_park )
```
