```python
import plotly.express as px
import pandas as pd
import numpy as np
import folium
import json
import geopandas as gpd
import folium.plugins
```

```python
data3 = pd.read_csv("읍별인구수.csv")
data3
```

```python
GwangYang = folium.Map(location=[34.9406968,127.6958882], zoom_start=11)
```

```python
gwangyang = gpd.read_file("20.광양시_행정경계(읍면동).geojson")
gwangyang
```

```python
choropleth = folium.Choropleth(
    geo_data=gwangyang,
    data=data3,
    columns=['읍면동', '인구수'],
    key_on='feature.properties.ADM_DR_NM',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=3,
    legend_name='인구수현황',
    highlight=True,
    smooth_factor=0).add_to(GwangYang)
```

```python
GwangYang.save('./1-3-인구수비교.html')
```
