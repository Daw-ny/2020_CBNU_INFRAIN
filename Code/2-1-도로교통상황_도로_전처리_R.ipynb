## 상세도로망-시간대별추정량 시각화를 위한 전처리

```r
library(dplyr)
library(reshape2)
library(rgdal)
```

```r
doro <- read.csv('11.광양시_평일_일별_시간대별__추정교통량_LV6.csv')
```

```r
str(doro)
```

```r
doro$도로코드<- substr(doro$상세도로망_링크ID,1,9)
doro$상하행선번호<- substr(doro$상세도로망_링크ID,10,11)
## 상세도로망_링크ID 에서 도로코드만 추출하기 위한 과정
```

```r
head(doro)
```

```r
doro %>% group_by(도로코드,읍면동명) %>%
            filter(시간적범위=="전일")
## 시간범위는 전일을 기준으로 뽑음
```

```r
doro_csv <- doro %>% group_by(도로코드,읍면동명,시간적범위,평일주말) %>%
                summarise(전체=sum(전체),승용차=sum(승용차),버스=sum(버스),화물차=sum(화물차)) %>%
                filter(시간적범위=="전일")
doro_csv
## 도로코드를 기준으로 통행하는 차량의 대수를 합치려는 과정
```

```r
write.csv(doro_csv,"도로.csv",row.names=F)
```

## 부록 : 차량 통행량의 3분할 기준

```r
doro_csv$전체 %>% quantile(probs=c(0,0.6,0.9,1))
```
