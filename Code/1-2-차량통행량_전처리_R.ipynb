## 평일 일별 추정 교통량

```r
library(dplyr)
library(reshape2)
library(rgdal)
```

```r
traffic <- read.csv('11.광양시_평일_일별_시간대별__추정교통량_LV6.csv',header=T)
traffic
```

```r
traffic <- traffic %>% group_by(읍면동명) %>%
                filter(시간적범위=="전일") %>%
                summarise(통행차량=sum(전체))
## 전체적인 통행량을 비교하기 위해 통행차량대수의 합을 구함.
```

```r
traffic
```

```r
write.csv(traffic,"평일일별추정교통량.csv",row.names=F)
```
